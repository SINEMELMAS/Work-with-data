***Eksik Veri Analizi***

```python
import numpy as np
import pandas as pd
V1 = np.array([1,3,6,np.NaN,7,1,np.NaN,9,15])
V2 = np.array([7,np.NaN,5,8,12,np.NaN,np.NaN,2,3])
V3 = np.array([np.NaN,12,5,6,14,7,np.NaN,2,31])
df= pd.DataFrame(
        {"V1" : V1,
         "V2" : V2,
         "V3" : V3}
)
df
```

```python
df.isnull().sum()
```

```python
df.notnull().sum()
```

```python
df.isnull().sum().sum()
```

```python
df.isnull()
```

```python
df[df.isnull().any(axis = 1)]
```

```python
df[df.notnull().all(axis = 1)]
```

```python
df[df["V1"].notnull() & df["V2"].notnull() & df["V3"].notnull()]
```

**Silme yöntemleri**

```python
import numpy as np
import pandas as pd
V1 = np.array([1,3,6,np.NaN,7,1,np.NaN,9,15])
V2 = np.array([7,np.NaN,5,8,12,np.NaN,np.NaN,2,3])
V3 = np.array([np.NaN,12,5,6,14,7,np.NaN,2,31])
df= pd.DataFrame(
        {"V1" : V1,
         "V2" : V2,
         "V3" : V3}
)
df
```

```python
df.dropna()
```

```python
df.dropna(how = "all")#tümü eksik olanı sil
```

```python
df.dropna(axis = 1)#içinde eksik olan tüm sütunları sil
```

```python
df.dropna(axis = 1, how = "all")#tüm değerlei NaN olan sütunu sil
```

```python
df
```

```python
df.dropna(axis = 0, how=  "all", inplace=True)
```

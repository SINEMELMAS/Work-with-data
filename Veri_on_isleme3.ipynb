```python
import numpy as np
import pandas as pd
V1 = np.array([1,3,6,np.NaN,7,1,np.NaN,9,15])
V2 = np.array([7,np.NaN,5,8,12,np.NaN,np.NaN,2,3])
V3 = np.array([np.NaN,12,5,6,14,7,np.NaN,2,31])
df= pd.DataFrame(
        {"V1" : V1,
         "V2" : V2,
         "V3" : V3}
)
df
```

```python
#sayısal değişkenler atama
```

```python
df['V1'].fillna(df['V1'].mean())
```

```python
df.apply(lambda x: x.fillna(x.mean()), axis=0)
```

```python
df.fillna(df.mean()[:])
```

```python
df.fillna(df.mean()["V1":"V3"])
```

```python
df["V3"].fillna(df["V3"].median())
```

```python
df.where(pd.notna(df), df.mean(), axis='columns')
```

**Kategorik Değişken Kırılımında Değer Atama**

```python
V1= np.array([1,3,6,np.NaN,7,1,np.NaN,9,15])
V2= np.array([7, np.NaN, 5, 8, 12, np.NaN, np.NaN, 2, 3])
V3= np.array([np.NaN, 12, 5, 6, 14, 7, np.NaN, 2, 31])
v4= np.array(["IT","IT","IK","IK","IK","IK","IK","IT","IT"], dtype=object)
df= pd.DataFrame(
        {"maas" : V1,
         "V2" : V2,
         "V3" : V3,
         "departman" : v4}
)
df
```

```python
df.groupby("departman")["maas"].mean()
```

```python
df["maas"].fillna(df.groupby("departman")["maas"].transform("mean"))
```

**Kategorik Değişkenler için Eksik Değer Atama**

```python
V1= np.array([1,3,6,np.NaN,7,1,np.NaN,9,15])
v4= np.array(["IT", np.NaN,"IK","IK","IK","IK","IK","IT","IT"], dtype=object)
df= pd.DataFrame(
        {"maas" : V1,
         "departman" : v4}
)
df
```

```python
df["departman"].fillna(df["departman"].mode()[0])
```

```python
df["departman"].fillna(method="bfill")#kendinden sonra gelenle doldurulur
```

```python
df['departman'].fillna(method='ffill')#kendinden önce gelenle doldurulur
```

```python
#random forest
import seaborn as sns
df= sns.load_dataset('titanic')
df= df.select_dtypes(include=['float64','int64'])
print(df.head())
df.isnull().sum()
```

```python
df.isnull().sum()
```

```python
var_names= list(df)
```

```python
n_df = np.array(df)
```

```python
from sklearn.experimental import enable_iterative_imputer  # Gerekli
from sklearn.impute import IterativeImputer

# IterativeImputer ile eksik değerleri doldurma
imputer = IterativeImputer(random_state=42)
dff = imputer.fit_transform(n_df)
```

```python
dff = pd.DataFrame(dff, columns=var_names)
```

```python
dff.isnull().sum()
```
